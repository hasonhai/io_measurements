# Methodology

__Tools__

For disk bandwidth test, we use Fio. Fio is short for Flexible IO, a versatile IO workload generator. It was written to benchmark or verify changes to the Linux IO subsystem by Linux developers. Fio is flexible enough to allow detailed workload setups, and it contains the necessary reporting to make sense of the data at completion. Fio is widely used as an industry standard benchmark, stress testing tool, and for IO verification purposes. Fio isn't tied to any OS. It works on any platform from Windows to HP-UX to Android. There are also native IO engines on Linux, Windows, Solaris, etc. This is a key feature to make the measurement repeatable across platforms.

Fio also supports three different types of output formats. The “classic” output is the default which dumps workload statistics at the end of the workload. There's also support for a CSV format called Terse, though that's slowly diminishing in favor of a JSON-based output format. The latter is far more flexible and has the advantage of being simple to parse for people and computers. In this report, we use the Terse format for the output.

__Test Design__

During our measurement, we saw that the bandwidth of random read/write and sequential read/write is very different. To keep the same setting for the test and control the time for each test, we have to consider carefully the size of transfered data, number of parallel threads, number of VMs on one host. If the size is too small, the job finish too quickly, we cannot see the a stable bandwidth of sequential read/write. If the data volume is too big, the random read/write test is take too long to finish. The number of parallel threads also affect a lot on the bandwith (as explained in the result assessment section). We have 5 different servers which have same specs. We do the same test on 5 servers to have a referrent for a good result. We check whether our result is usable by examine at the same time the result from 5 servers. If the result on one server is different from the others, we can conclude that our data were affected too much by other users or processes. On each server, we run same number of Virtual Machines. We run Fio on the physical host with different number of Fio threads (from 1 to 4 concurrent threads). Then with the same configuration, we run it from inside the VMs. We also make the comparison with the case that each VM has only one thread but we increase the number of concurrent running VMs on the same host. We also use identical Operating Systems and identical Os setting across platforms to make sure that the measurement result is relisable. 

Our configuration file for Fio has this structure:

    ; sample setting of one thread
    [sequential-read]
    rw=read
    size=(512mb/number_of_parallel_threads)
    directory=/home/cloud-user/fiodata
    ioengine=libaio
    direct=1
    iodepth=1/16/32

Where:
* ___rw___ defines the kind of disk access pattern: read (sequential reads), write (sequential writes), randread (random reads), or randwrite (random writes). We specially focus on these 4 different access patterns. In reality, HDFS only have sequential read and sequential write. But in case of the physical host, it may see the pattern as random read or random write because of the inter-switching between VMs on the same host or between threads on the same server. 
* ___size___ is the total size of file io for this job. Fio will run until this many bytes has been transferred. We choose the total transfer volume to be 512 MBs. This is long enough to... bla bla bla. The data will be divided to threads equally between parallel threads:
    + 1 thread: transfers entire 512 MBs
    + 2 threads: each thread transfer 256 MBs
    + 3 threads: each thread transfer 170 MBs
    + 4 threads: each thread transfer 128 MBs
* ___direct I/O___ uses the normal filesystem access, but disables buffering and locking operations
* ___iodepth___ defines the I/O depth of the system. This defines how many io units to keep in flight against the file. The default is 1 for each file defined in this job, can be overridden with a larger value for higher concurrency. We use I/O depth of 1, 16 and 32. Default setting in some of the Linux distribution set I/O depth to 256 or 512.
* ___ioengine___ is libaio which indicates the application to use AIO subsystem. AIO (Asynchronous Linux IO) help to overlaps processing with I/O Operations. Application can submit (batch) IO without waiting for completion. This benefits physical IO in particular when prefetching data. Besides, It allows to accumulate read or write requests so that the IO subsystem can optimize performance by grouping or reordering the IO request in favor of sequential access and large request. It helps to separate calls for submission and completion indication and pipelines operations to improve throughput. It also helps to improved utilization of CPU and devices.

```python
# To calculate the average throughput, we first calculate average throughput each thread has in one test, then again take the
# average of 5 tests
avgtpin1test = onehostdata.T.mean() # Average throughput of each thread/VM in one test
avgtpin5tests = avgtpin1test.groupby(level=0).mean() # Average throughput of each thread/VM in 5 tests
avgtpin5tests = avgtpin5tests / 1024 # Converting to Megabytes/second
```

```python
# Putting average throughput per thread/VM to vector before plotting
bwsrbf = avgtpin5tests.filter(regex="^SRBFTH").tolist()
bwswbf = avgtpin5tests.filter(regex="^SWBFTH").tolist()
bwrrbf = avgtpin5tests.filter(regex="^RRBFTH").tolist() 
bwrwbf = avgtpin5tests.filter(regex="^RWBFTH").tolist()
bwsrsvm = avgtpin5tests.filter(regex="^SR1VMTH").tolist()
bwswsvm = avgtpin5tests.filter(regex="^SW1VMTH").tolist()
bwrrsvm = avgtpin5tests.filter(regex="^RR1VMTH").tolist() 
bwrwsvm = avgtpin5tests.filter(regex="^RW1VMTH").tolist()
bwsrmvm = avgtpin5tests.filter(regex="^SR[1-4]VMTH1").tolist()
bwswmvm = avgtpin5tests.filter(regex="^SW[1-4]VMTH1").tolist()
bwrrmvm = avgtpin5tests.filter(regex="^RR[1-4]VMTH1").tolist() 
bwrwmvm = avgtpin5tests.filter(regex="^RW[1-4]VMTH1").tolist()
```

```python
# Calculate aggregated throughput on each host of each test, then taking the average
aggtpin1test = onehostdata.T.sum() # Aggregate throughput on each host in one test
avgaggtpintestgrouped = aggtpin1test.groupby(level=0)
avgaggtpin5tests = avgaggtpintestgrouped.mean() # Then calculate the average of the total throughput through 5 tests
avgaggtpin5tests = avgaggtpin5tests / 1024 # Convert to MB/s
ciaggtpin5tests = avgaggtpintestgrouped.aggregate(lambda x: np.std(x) / np.sqrt(x.count()) * 1.96) # Confident interval of 95%
ciaggtpin5tests = ciaggtpin5tests / 1024 # Convert to same scale with mean
```

```python
# Putting average aggregate throughput to vector before plotting
aggbwsrbf = avgaggtpin5tests.filter(regex="^SRBFTH").tolist()
aggbwswbf = avgaggtpin5tests.filter(regex="^SWBFTH").tolist()
aggbwrrbf = avgaggtpin5tests.filter(regex="^RRBFTH").tolist() 
aggbwrwbf = avgaggtpin5tests.filter(regex="^RWBFTH").tolist()
aggbwsrsvm = avgaggtpin5tests.filter(regex="^SR1VMTH").tolist()
aggbwswsvm = avgaggtpin5tests.filter(regex="^SW1VMTH").tolist()
aggbwrrsvm = avgaggtpin5tests.filter(regex="^RR1VMTH").tolist() 
aggbwrwsvm = avgaggtpin5tests.filter(regex="^RW1VMTH").tolist()
aggbwsrmvm = avgaggtpin5tests.filter(regex="^SR[1-4]VMTH1").tolist()
aggbwswmvm = avgaggtpin5tests.filter(regex="^SW[1-4]VMTH1").tolist()
aggbwrrmvm = avgaggtpin5tests.filter(regex="^RR[1-4]VMTH1").tolist() 
aggbwrwmvm = avgaggtpin5tests.filter(regex="^RW[1-4]VMTH1").tolist()

# Confidence Interval
ciaggbwsrbf = ciaggtpin5tests.filter(regex="^SRBFTH").tolist()
ciaggbwswbf = ciaggtpin5tests.filter(regex="^SWBFTH").tolist()
ciaggbwrrbf = ciaggtpin5tests.filter(regex="^RRBFTH").tolist() 
ciaggbwrwbf = ciaggtpin5tests.filter(regex="^RWBFTH").tolist()
ciaggbwsrsvm = ciaggtpin5tests.filter(regex="^SR1VMTH").tolist()
ciaggbwswsvm = ciaggtpin5tests.filter(regex="^SW1VMTH").tolist()
ciaggbwrrsvm = ciaggtpin5tests.filter(regex="^RR1VMTH").tolist() 
ciaggbwrwsvm = ciaggtpin5tests.filter(regex="^RW1VMTH").tolist()
ciaggbwsrmvm = ciaggtpin5tests.filter(regex="^SR[1-4]VMTH1").tolist()
ciaggbwswmvm = ciaggtpin5tests.filter(regex="^SW[1-4]VMTH1").tolist()
ciaggbwrrmvm = ciaggtpin5tests.filter(regex="^RR[1-4]VMTH1").tolist() 
ciaggbwrwmvm = ciaggtpin5tests.filter(regex="^RW[1-4]VMTH1").tolist()
```

## CPU Utilization

For CPU utilization, FIO relies on _getrusage()_. _Getrusage()_ returns resource usage statistics for the calling thread.
The resource usages are returned in the structure pointed to by usage, which has the following form:
~~~
struct rusage {
    struct timeval ru_utime; /* user CPU time used */
    struct timeval ru_stime; /* system CPU time used */
    ...
    long   ru_minflt;        /* page reclaims (soft page faults) */
    long   ru_majflt;        /* page faults (hard page faults) */
    long   ru_nswap;         /* swaps */
    ...
    long   ru_nvcsw;         /* voluntary context switches */
    long   ru_nivcsw;        /* involuntary context switches */
};
~~~
and contains inside the _struct_:
* _ru_utime_: This is the total amount of time spent executing in user mode, expressed in a _timeval_ structure (seconds plus microseconds).
* _ru_stime_: This is the total amount of time spent executing in kernel mode, expressed in a _timeval_ structure (seconds plus microseconds).
* _ru_nvcsw_: The number of times a context switch resulted due to a process voluntarily giving up the processor before its time slice was completed (usually to await availability of a resource).
* _ru_nivcsw_: The number of times a context switch resulted due to a higher priority process becoming runnable or because the current process exceeded its time slice.

FIO reports the CPU utilization for each thread by (_user_time/total_run_time_) and (_system_time/total_run_time_), where:
* _user_time_ is the duration that the thread spending in user space
* _system_time_ is the duration the t
* _total_run_time_ is the duration from the time that thread starts until the time that thread finishs.

~~~
	runtime = ts->total_run_time;
	if (runtime) {
		double runt = (double) runtime;

		usr_cpu = (double) ts->usr_time * 100 / runt;
		sys_cpu = (double) ts->sys_time * 100 / runt;
	} else {
		usr_cpu = 0;
		sys_cpu = 0;
	}
~~~
In this report, we especially forcus on the the serving time (the time that thread actually uses the CPU) and the waiting time (the time that thread has to wait for the other when its slice is up).

## IOPS

```python
iopsresult = pd.read_csv('data/dtIOPS.csv', delimiter=' ', header=None)
iopsdata = iopsresult[range(1,21)]
iopsindex = iopsresult[0]
iopsdata.index = iopsindex.str.replace('T1|T2|T3|T4|T5', '',case=False)
```

```python
onehostdata = iopsdata[selectedhost]
```

```python
# To calculate the average throughput, we first calculate average throughput each thread has in one test, then again take the
# average of 5 tests
avgiopsin1test = onehostdata.T.mean() # Average throughput of each thread/VM in one test
avgiopsin5tests = avgiopsin1test.groupby(level=0).mean() # Average throughput of each thread/VM in 5 tests
```

```python
# Putting average throughput per thread/VM to vector before plotting
iopssrbf = avgiopsin5tests.filter(regex="^SRBFTH").tolist()
iopsswbf = avgiopsin5tests.filter(regex="^SWBFTH").tolist()
iopsrrbf = avgiopsin5tests.filter(regex="^RRBFTH").tolist() 
iopsrwbf = avgiopsin5tests.filter(regex="^RWBFTH").tolist()
iopssrsvm = avgiopsin5tests.filter(regex="^SR1VMTH").tolist()
iopsswsvm = avgiopsin5tests.filter(regex="^SW1VMTH").tolist()
iopsrrsvm = avgiopsin5tests.filter(regex="^RR1VMTH").tolist() 
iopsrwsvm = avgiopsin5tests.filter(regex="^RW1VMTH").tolist()
iopssrmvm = avgiopsin5tests.filter(regex="^SR[1-4]VMTH1").tolist()
iopsswmvm = avgiopsin5tests.filter(regex="^SW[1-4]VMTH1").tolist()
iopsrrmvm = avgiopsin5tests.filter(regex="^RR[1-4]VMTH1").tolist() 
iopsrwmvm = avgiopsin5tests.filter(regex="^RW[1-4]VMTH1").tolist()
```

```python
# Calculate aggregated throughput on each host of each test, then taking the average
aggiopsin1test = onehostdata.T.sum() # Aggregate throughput on each host in one test
avgaggiopsintestgrouped = aggiopsin1test.groupby(level=0)
avgaggiopsin5tests = avgaggiopsintestgrouped.mean() # Then calculate the average of the total throughput through 5 tests
ciaggiopsin5tests = avgaggiopsintestgrouped.aggregate(lambda x: np.std(x) / np.sqrt(x.count()) * 1.96) # Confident interval of 95%
```

```python
# Putting average aggregate throughput to vector before plotting
aggiopssrbf = avgaggiopsin5tests.filter(regex="^SRBFTH").tolist()
aggiopsswbf = avgaggiopsin5tests.filter(regex="^SWBFTH").tolist()
aggiopsrrbf = avgaggiopsin5tests.filter(regex="^RRBFTH").tolist() 
aggiopsrwbf = avgaggiopsin5tests.filter(regex="^RWBFTH").tolist()
aggiopssrsvm = avgaggiopsin5tests.filter(regex="^SR1VMTH").tolist()
aggiopsswsvm = avgaggiopsin5tests.filter(regex="^SW1VMTH").tolist()
aggiopsrrsvm = avgaggiopsin5tests.filter(regex="^RR1VMTH").tolist() 
aggiopsrwsvm = avgaggiopsin5tests.filter(regex="^RW1VMTH").tolist()
aggiopssrmvm = avgaggiopsin5tests.filter(regex="^SR[1-4]VMTH1").tolist()
aggiopsswmvm = avgaggiopsin5tests.filter(regex="^SW[1-4]VMTH1").tolist()
aggiopsrrmvm = avgaggiopsin5tests.filter(regex="^RR[1-4]VMTH1").tolist() 
aggiopsrwmvm = avgaggiopsin5tests.filter(regex="^RW[1-4]VMTH1").tolist()

# Confidence Interval
ciaggiopssrbf = ciaggiopsin5tests.filter(regex="^SRBFTH").tolist()
ciaggiopsswbf = ciaggiopsin5tests.filter(regex="^SWBFTH").tolist()
ciaggiopsrrbf = ciaggiopsin5tests.filter(regex="^RRBFTH").tolist() 
ciaggiopsrwbf = ciaggiopsin5tests.filter(regex="^RWBFTH").tolist()
ciaggiopssrsvm = ciaggiopsin5tests.filter(regex="^SR1VMTH").tolist()
ciaggiopsswsvm = ciaggiopsin5tests.filter(regex="^SW1VMTH").tolist()
ciaggiopsrrsvm = ciaggiopsin5tests.filter(regex="^RR1VMTH").tolist() 
ciaggiopsrwsvm = ciaggiopsin5tests.filter(regex="^RW1VMTH").tolist()
ciaggiopssrmvm = ciaggiopsin5tests.filter(regex="^SR[1-4]VMTH1").tolist()
ciaggiopsswmvm = ciaggiopsin5tests.filter(regex="^SW[1-4]VMTH1").tolist()
ciaggiopsrrmvm = ciaggiopsin5tests.filter(regex="^RR[1-4]VMTH1").tolist() 
ciaggiopsrwmvm = ciaggiopsin5tests.filter(regex="^RW[1-4]VMTH1").tolist()
```

```python
x_axis = [1,2,3,4]

plt.subplot(241)
plt.plot(x_axis, iopssrbf, x_axis, iopssrsvm, x_axis, iopssrmvm)
plt.axis([0.9, 4.1, 0, 40000])
plt.grid(True)
plt.title('Sequential Read')
pylab.xlabel('concurrent threads/VMs')
pylab.ylabel('IOPS')

plt.subplot(242)
plt.plot(x_axis, iopsswbf, x_axis, iopsswsvm, x_axis, iopsswmvm)
plt.axis([0.9, 4.1, 0, 40000])
plt.grid(True)
plt.title('Sequential Write')
pylab.xlabel('concurrent threads/VMs')
pylab.ylabel('IOPS')

plt.subplot(243)
plt.plot(x_axis, iopsrrbf, x_axis, iopsrrsvm, x_axis, iopsrrmvm)
plt.axis([0.9, 4.1, 0, 40000])
plt.grid(True)
plt.title('Random Read')
pylab.xlabel('concurrent threads/VMs')
pylab.ylabel('IOPS)')

plt.subplot(244)
plt.plot(x_axis, iopsrwbf, label = 'bare-metal IO request')
plt.plot(x_axis, iopsrwsvm, label = 'Avg. IO requests per thread SVM-MT')
plt.plot(x_axis, iopsrwmvm, label = 'Avg. IO requests per VM MVM-ST')
plt.axis([0.9, 4.1, 0, 40000])
plt.grid(True)
plt.title('Random Write')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
pylab.xlabel('concurrent threads/VMs')
pylab.ylabel('IOPS')

plt.subplot(245)
plt.errorbar(x_axis, aggiopssrbf, ciaggiopssrbf)
plt.errorbar(x_axis, aggiopssrsvm, ciaggiopssrsvm)
plt.errorbar(x_axis, aggiopssrmvm, ciaggiopssrmvm)
plt.axis([0.9, 4.1, 0, 40000])
plt.grid(True)
plt.title('Sequential Read')
pylab.xlabel('concurrent threads/VMs')
pylab.ylabel('IOPS')

plt.subplot(246)
plt.errorbar(x_axis, aggiopsswbf, ciaggiopsswbf)
plt.errorbar(x_axis, aggiopsswsvm, ciaggiopsswsvm)
plt.errorbar(x_axis, aggiopsswmvm, ciaggiopsswmvm)
plt.axis([0.9, 4.1, 0, 40000])
plt.grid(True)
plt.title('Sequential Write')
pylab.xlabel('concurrent threads/VMs')
pylab.ylabel('IOPS')

plt.subplot(247)
plt.errorbar(x_axis, aggiopsrrbf, ciaggiopsrrbf)
plt.errorbar(x_axis, aggiopsrrsvm, ciaggiopsrrsvm)
plt.errorbar(x_axis, aggiopsrrmvm, ciaggiopsrrmvm)
plt.axis([0.9, 4.1, 0, 40000])
plt.grid(True)
plt.title('Random Read')
pylab.xlabel('concurrent threads/VMs')
pylab.ylabel('IOPS')

plt.subplot(248)
plt.errorbar(x_axis, aggiopsrwbf, ciaggiopsrwbf, label = 'bare-metal IO requests')
plt.errorbar(x_axis, aggiopsrwsvm, ciaggiopsrwsvm, label = 'avg. agg. IOPS SVM-MT')
plt.errorbar(x_axis, aggiopsrwmvm, ciaggiopsrwmvm, label = 'avg. agg. IOPS MVM-ST')
plt.axis([0.9, 4.1, 0, 40000])
plt.grid(True)
plt.title('Random Write')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
pylab.xlabel('concurrent threads/VMs')
pylab.ylabel('IOPS')
```
